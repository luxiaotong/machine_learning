```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(1)

alpha = 0.001
m = 100
X = np.arange(100)
y = X + np.random.random(m) * 5


print(y)
```

```python
W = 1
b = 0
y1 = W * X + b
plt.plot(X, y, 'o')
plt.plot(X, y1)
```

```python
W = 1
b = 0

for i in range(0, 100):
    y_pred = W * X[i] + b
    J_cost = (1/(2*m)) * np.sum(np.square(y_pred-y[i]))
    print("J_cost:", J_cost)
    
    
    W_delta = (alpha/m) * np.sum(np.multiply((y_pred - y[i]), X[i]))
    b_delta = (alpha/m) * np.sum(y_pred - y[i])
    
    W -= W_delta
    b -= b_delta
```

```python
y_rst = W * X + b
#plt.plot(X, y, X, y_rst)
plt.plot(X, y, 'o')
plt.plot(X, y_rst)
```
