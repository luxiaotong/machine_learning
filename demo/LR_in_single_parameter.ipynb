```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(1)

alpha = 0.001
m = 100

X = np.arange(m)
y = X + np.random.random(m) * 5
```

```python
W = 1
b = 0
y1 = W * X + b
plt.plot(X, y, 'o')
plt.plot(X, y1)
```

```python
W = 1
b = 0

for epoch in range(0, 1):
    for i in range(0, 100):
        y_pred = W * X[i] + b
        print("X:", X[i])
        print("W:", W)
        print("b:", b)
        print("y_pred:", y_pred)
        print("y:", y[i])
#        J_cost = (1/2) * np.sum(np.square(y_pred-y[i]))
        J_cost = (1/2) * np.square(y_pred-y[i])
        print("J_cost:", J_cost)

    
#        W_delta = alpha * np.sum(np.multiply((y_pred - y[i]), X[i]))
#        b_delta = alpha * np.sum(y_pred - y[i])

        W_delta = alpha * np.multiply((y_pred - y[i]), X[i])
        b_delta = alpha * (y_pred - y[i])
        print("W_delta:", W_delta)
        print("b_delta:", b_delta)

        W -= W_delta
        b -= b_delta
        print("W:", W)
        print("b:", b)
        print()
```
