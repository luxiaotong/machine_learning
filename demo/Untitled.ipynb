```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(1)

alpha = 0.1
m = 100
W = np.ones(m)
b = np.ones(m)
X = np.ones(m)
y = np.random.random(m)
```

```python
plt.plot(y)
```

```python
for i in range(1, 100):
    y_pred = W.T * X + b
    J_cost = (1/(2*m)) * np.sum(np.square(y_pred-y))
    
    W_delta = (alpha/m) * np.sum(np.multiply((y_pred - y), X))
    b_delta = (alpha/m) * np.sum(y_pred - y)
    
    W -= W_delta
    b -= b_delta
    
    print(W)
    print(b)
    
print("test:")
print(W)
print(b)
```
